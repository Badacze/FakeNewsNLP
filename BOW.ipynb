<h1>Tensorflow

```python
from tensorflow import keras
from typing import List
from tensorflow.keras.preprocessing.text import Tokenizer
import pandas as pd

cols = ['json','label','content','nie','chce','mi','sie','zgadywac','co','ozancza','kazda','kolumna','wiec','skip']

df = pd.read_csv('train_filtered.tsv', sep='\t', names=cols)

def bowTensorflow(content) -> List[dict]:
    bow_list = []
    tokenizer = Tokenizer()
    tokenizer.fit_on_texts(content.tolist())
    sequences = tokenizer.texts_to_sequences(content) # reprezentajca słowa jako liczba
    word_index = tokenizer.word_index # wszystkie unikalne słowa w tekście jako indeksy

    for i, sentence in enumerate(content):
        bow_dict = {}
        current_sequence = sequences[i]

        for word, idx in word_index.items(): # zliczanie słów w każdej komórce
            count = current_sequence.count(idx)
            bow_dict[word] = count

        bow_list.append(bow_dict)

    return bow_list, word_index # lista słwoników ze słowami dla każdej wypowiedzi + indeksy

print(df['content'][0])
```

```python
LIAR_BOW, indexes = bowTensorflow(df['content'])
```

```python
LIAR_BOW_dataframe = pd.DataFrame(data=LIAR_BOW, columns=indexes)

LIAR_BOW_dataframe
```

<h1>Sklearn

```python
from sklearn.feature_extraction.text import CountVectorizer

def bowSklearn(frame):
  vectorizer = CountVectorizer()
  X = vectorizer.fit_transform(frame) # macierz z występowaniami
  words = vectorizer.get_feature_names_out() # wszystkie słowa

  sk_bow = pd.DataFrame(data=X.toarray(), columns=words) # od razu jako data frame | mniej do pisania i szybciej działa

  return sk_bow

sk_bow = bowSklearn(df['content'])
```
